```python
import pandas as pd
from db_util import Database
```

```python
db = Database("C:\\Users\\rritz\\Desktop\\Track Results")
```

```python
year = 2024
file = str(year) + "_IHSAA.csv"
```

```python
# load schools and enrollments#
# For any new schools that are added, populate the long/lat and avg home value fields manually

df = pd.read_csv(file)

for index, row in df.iterrows():
    name = row["SchoolName"]
    
    school_id = db.get_school_id(name)

    if school_id == None:
        type = row["School Type"]
        nickname = row["Nickname"]
        address = row["Address1"];
        city = row["City"]
        zip = row["ZipCode"]
        db.insert_school(name, "", type, nickname, address, city, zip)
        school_id = db.get_school_id(name)
    
    db.insert_school_enrollment(school_id, year, row["Enrollment"])
```
